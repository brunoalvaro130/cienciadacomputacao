```
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(0.0,4.0,0.01)
y = 1 / 1 + (np.exp(-x))

fig, ax = plt.subplots()

ax.plot(x, y)

ax.grid()

plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(0.0,4.0,0.01)
y = (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

fig, ax = plt.subplots()

ax.plot(x, y)

ax.grid()

plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(0.0,4.0,0.01)
y = x**2 - x - 2

fig, ax = plt.subplots()

ax.plot(x, y)

ax.grid()

plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(0.0,4.0,0.01)
y = x + (np.sin(x)**2) / (np.cos(x))*np.pi*(np.sin(x))

fig, ax = plt.subplots()

ax.plot(x, y)

ax.grid()

plt.show()
```
